```
######## Prediction table ##########################33


avgAcc=dict()
predList=[]
for df in ["os","is","ws"]:
    print("|||||||||||||||{}|||||||||||||||||".format(df))
    avgAcc[df]=[]
    for rlimit in [500,550,650,700,750]: 
        layerAvg=0   
        for layeri in range(0,11):
            print("---------------------{}--------------------".format(rlimit))    
            reuseLimit = rlimit #Values that should be cached
            predBits=10
            predBitsMask=(1<<predBits)-1
            counterSize=3
            predOff=2
            ghrBits=10
            ghrBitsMask=(1<<ghrBits)-1
            predTable=dict()
            for i in range(1<<predBits):
                predTable[i]=0


            lay="/content/drive/MyDrive/test_run/scale_example_run_128x128_{}/layer{}/IFMAP_DRAM_TRACE.csv".format(df,layeri)
            # reuseDist = dict()
            reuseDist=[]
            addressCycle=dict()
            with open(lay) as fst0:
                datareader = csv.reader(fst0)
                rowidx=0
                for row in datareader:
                    # reuseDist[rowidx]=[]
                    for cell in range(1,len(row)):
                        addr = int(float(row[cell]))
                        if addr not in addressCycle.keys():
                            addressCycle[addr]=rowidx
                            # reuseDist.append([addr,False]) #set to some high value so it isnt used to predict
                        else:
                            reuseDist.append([addr,(rowidx-addressCycle[addr])<=reuseLimit])
                            addressCycle[addr]=rowidx
                        
                    rowidx+=1

            def predict(addr,ghr=None):
                #True if cache else False
                if ghr:
                    addr=addr&(predBitsMask<<predOff)
                    ghr=ghr&ghrBitsMask
                    if predTable[(addr^ghr)%(predBitsMask+1)] > ((1<<counterSize)/2) -1:
                        return True
                    else:
                        return False
                else:
                    addr=addr&(predBitsMask<<predOff)
                    if predTable[(addr)%(predBitsMask+1)] > ((1<<counterSize)/2) -1:
                        return True
                    else:
                        return False

                    
            def train(prediction,label,addr,ghr=None):
                if ghr:
                    addr=addr&(predBitsMask<<predOff)
                    ghr=ghr&ghrBitsMask
                    if (prediction and not label):
                        predTable[(addr^ghr)%(predBitsMask+1)]-=1
                    elif (not prediction and label):
                        predTable[(addr^ghr)%(predBitsMask+1)]+=1
                    elif (prediction and label) and predTable[(addr^ghr)%(predBitsMask+1)]<(1<<counterSize)-1:
                        predTable[(addr^ghr)%(predBitsMask+1)]+=1
                    elif predTable[(addr^ghr)%(predBitsMask+1)]>0:
                        predTable[(addr^ghr)%(predBitsMask+1)]-=1
                else:
                    addr=addr&(predBitsMask<<predOff)
                    if (prediction and not label):
                        predTable[(addr)%(predBitsMask+1)]-=1
                    elif (not prediction and label):
                        predTable[(addr)%(predBitsMask+1)]+=1
                    elif (prediction and label) and predTable[(addr)%(predBitsMask+1)]<(1<<counterSize)-1:
                        predTable[(addr)%(predBitsMask+1)]+=1
                    elif predTable[(addr)%(predBitsMask+1)]>0:
                        predTable[(addr)%(predBitsMask+1)]-=1
            ghr=0
            def updateGhr(ghr,label):
                ghr=(ghr<<1)&ghrBitsMask
                if label:
                    ghr+=1
                return ghr

            TP=0
            TN=0
            FP=0
            FN=0
            for rowidx in range(len(reuseDist)):
                cell = reuseDist[rowidx]
                addr=cell[0]
                label=cell[1]
                prediction=predict(addr)
                if label and prediction:
                    TP+=1
                elif not label and not prediction:
                    TN+=1
                elif label and not prediction:
                    FN+=1
                else:
                    FP+=1
                train(prediction,label,addr)
                # ghr = updateGhr(ghr,label)

            # print(TP,TN,FP,FN)
            print("Confusion:")
            print("Prediction")
            print("G ",TP,FN)
            print("T ",FP,TN)
            print("ACC: ",((TN+TP)/(TN+TP+FN+FP))*100)
            layerAvg+=(TN+TP)/(TN+TP+FN+FP)
            predList.append(predTable)
        avgAcc[df].append(layerAvg/11)
```
