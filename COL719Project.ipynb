#**Linking the inputs**

SCALE-Sim uses two inputs to work

1.   First is a config file like `scale.cfg` below. The config file describes the architecture, dataflow and a few modes for the tool to run.


2.   Second is a CSV file, which specifies workloads as neural network layer parameters. In our example, this file is the `alexnet_part.csv`, which captures the description of the first convolution layer in Alexnet.



```
from scalesim.scale_sim import scalesim

content_path = "/content/drive/MyDrive/scale-sim-v2-main"
config = content_path + "/configs/scale.cfg"
topo = content_path + "/topologies/conv_nets/alexnet_full.csv"
```

#**Instantiating the simulator object**

Here we create an object of the SCALE-Sim calls, and pass the paths to input files. We also set the tool to run in a verbose mode and signal it to generate trace files as well (save_disk_space=True, will suppress trace generation)

```
#OS
top = "/content/drive/MyDrive/test_run"
s = scalesim(save_disk_space=True, verbose=True,
              config=config,
              topology=topo
              )
```

#**Run Simulation**

That's it, now we run the simulation with the following call. 

```
s.run_scale(top_path=top)
```

```
# import pandas as pd

# df = pd.read_csv("scale_example_run_32x32_os/layer0/FILTER_SRAM_TRACE.csv")

import csv
import numpy as np
%matplotlib inline
#The line above is necesary to show Matplotlib's plots inside a Jupyter Notebook
from matplotlib import pyplot as plt

def reuseHist(fp="/content/drive/MyDrive/test_run/scale_example_run_32x32_os_16/layer{}/IFMAP_DRAM_TRACE.csv"):
    # reuseDist = dict()

    # plt.figure(figsize=(20,15))
    fig, axis = plt.subplots(5,2,figsize=(25, 15))
    for lay in range(1,11):
        reuseDist = dict()
        fpLayer =fp.format(lay)
        with open(fpLayer) as fst0:
            datareader = csv.reader(fst0)
            rowidx=0
            for row in datareader:
                for cell in range(1,len(row)):
                    cell=int(cell)
                    if row[cell] not in reuseDist.keys():
                        reuseDist[row[cell]]=[]
                        reuseDist[row[cell]].append(rowidx)
                    else:
                        reuseDist[row[cell]].append(rowidx)
                    rowidx+=1


        # print(reuseDist.keys())
        count=0
        dists=[]
        for i in reuseDist.keys():
            
            for j in range(len(reuseDist[i])-1):
                dists.append(reuseDist[i][j+1]-reuseDist[i][j])
            # print("address {}:".format(i),np.mean(dists),np.min(dists),np.max(dists),np.std(dists))
            # print("Address {}".format(i),dists)
            # print(reuseDist[i])
            # count+=1
            # if count > 1000:
            #     break

        if lay<6:
            axis[(lay-1),0].hist(dists,log=True)
            axis[(lay-1),0].set_title(str(lay)+"th layer")
        else:
            axis[(lay-1)%5,1].hist(dists,log=True)
            axis[(lay-1)%5,1].set_title(str(lay)+"th layer")

    plt.show()
```
